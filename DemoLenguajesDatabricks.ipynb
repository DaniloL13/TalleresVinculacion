## 2. Modelo de Regresión Logística

A través del modelo de Regresión Logística modelamos la **probabilidad** de observar cierto estado en la variable de respesta (variable a predecir) dada cierta información relacionada (predictores) en ```X```.
La expresión a continuación modela la probabilidad de la variable a predecir ```Y``` sea equivalente a 1 dada la información de sus predictores en ```X```:

$$ p(X) = Pr(Y = 1 | X)$$

Para modelar ```p(X)``` necesitamos utilizar una función que entregue valores de probabilidad entre 0 y 1 para todos los valores de ```X```. Una de las funciones que cumple este requerimiento es la **Función Logística**, especificada a continuación:

$$\large p(X) = \frac{e ^ {\beta_0 + \beta_1X}}{1 \; + \; e ^ {\beta_0 + \beta_1X}}$$

Manipulando la ultima ecuación, tenemos: 

$$\large \text{Odds} = \frac{p(X)}{1-p(X)} = e ^ {\beta_0 + \beta_1X} \;\;\;;\;\;\; \space \text{Odds} \in ({0,\inf})$$

Tomando el logaritmo a la expresión anterior, se define el ```log-Odds``` conocido también como ```logit```.

$$\large \text{log-Odds}=\log{(\frac{p(X)}{1-p(X)})} = \beta_0 + \beta_1X$$

A partir de esta ultima expresión, podemos apreciar que el modelo de Regresión Logística tienen una expresión lineal en ```X``` (a través de la relación ```log-Odds```): <br><br>

$$\large p(X) = \beta_0 + \beta_1X$$


**NOTAS:**<br>
Aunque la última ecuación representa una relación lineal en ```X```, la relación entre ```p(X)``` y ```X``` NO es una línea recta. Esto implica que el incremento en una unidad de ```X``` NO provoca un cambio en el ```log-odds``` de ```beta_1``` unidades, ni tampoco este cambio equivale a multiplicar el ```odds``` por el factor de ```e^beta_1```. <br> <br>
Lo que SI podemos afirmar es que si ```beta_1``` es positivo, entonces un incremento en ```X``` produce también un incremento en ```p(X)```. Lo homólogo sucede cuando ```beta_1``` es negativo, asociando el decremento de ```X``` con un decremento en ```p(X)```.

## 3. Dataset de Trabajo

**defaultData:** Información sobre caracteristicas de consumo de usuarios de tarjetas de crédito con el fin de predecir los usuarios que dejarán de pagar su balance pendiente (default = Yes). 

**Objetivo de Clasificación:** Predecir los usuarios que DEJARÁN DE PAGAR sus balances pendientes de consumos con tarjetas de crédito (**default = Yes**) y los que pagarán cumplidamente (**default = No**).

$$Pr (\text{default} = \text{Yes}| \text{student, balance, income})= Pr(y = 1|X) $$
$$p(X) = Pr(y = 1|X)$$

````python
#https://storage.googleapis.com/datasets-academy/Default.csv

defaultData = pd.read_csv('https://raw.githubusercontent.com/handytecmobi/datasets-academy/master/Default.csv')
defaultData.head()
````

## 4. Pre-procesamiento de datos

### 4.1) Manejo de variables categóricas


Las variables categóricas como ```default``` o ```student``` requieren una transformación de sus estados **YES , NO** a valores numéricos asociados que permitan utilizar esta información en los algoritmos.<br>
Lo ideal es asignar estos valores numéricos sin añadir sesgo (**bias**) a los datos. Las variables que se utilizan para realizar estas transformaciones de datos categóricos/estados a datos numéricos se conocen también como ```dummy variables``` o variables ficticias. <br>
Cuando manejamos 2 estados categóricos, este procedimiento se conoce también como técnicas de ```Binarización```. <br> <br>

En nuestro ejercicio con ```defaultData```, transformaremos los estados **Yes** y **No** a valores de **0 y 1**, respectivamente. El proceso lo realizaremos a través de un mapeo de datos utilizando un diccionario que especifica los valores dummy/ficticios que se asignará a las variables categóricas.

````python
dummyDict = {'No':0, 'Yes':1}
defaultData['default'] = defaultData['default'].map(dummyDict)
defaultData['student'] = defaultData['student'].map(dummyDict)
defaultData.head()
````

Ahora que los datos de las columnas ```default``` y ```student``` son numéricas, podemos visualizar cuántos clientes han dejado de pagar sus saldos pendientes (default = Yes) y cuántos continúan pagando con normalidad (default = No)

````python
print('Full Default Yes = ', sum(defaultData['default']== 1))
print('Full Default No = ' , sum(defaultData['default']== 0))
````

### 4.2) Selección de los datos para ```X , y```:

````python
X = defaultData[['student','balance','income']].values
y = defaultData['default'].values

print(X.shape, y.shape)
````

### 4.3) Normalización de los datos

Debido a que las diferentes variables predictoras pueden tener datos en diferentes rangos, es necesario aplicar una **técnica de normalización a los datos de los predictores** para que éstos se puedan procesar en escalas de valores comparables entre sí. Generalmente luego de normalizar los datos, el rango de valores de las variables predictoras va del 0 al 1. <br><br>
Las técnicas de normalización más comunes son:
- Normalización Estándar o Gaussiana:
$$XScaled = \frac{X-\mu}{\sigma}$$

- Normalización Min-Max:
$$XScaled = \frac{X- min(X)}{max(X)}$$
    
Para el caso de nuestro ejemplo, utilizaremos la Normalización Estándar: <br>

**NOTA:** Únicamente normalizamos los datos de las variables predictoras ```X```.

````python
scaler = StandardScaler()
XScaled = scaler.fit_transform(X)

aux = np.concatenate((X,XScaled),axis=1)
df = pd.DataFrame(aux, columns=['student','balance','income','student_S','balance_S','income_S'])
print(df.head())
````

## 5) Desarrollo del Modelo

### 5.1 ) División en Train y Test:

**```random_sate```** <br>
- Este parámetro inicializa el generador aleatoreo de números que indicará los índices de división los datos en las porciones de Train y Test. 
- Para obtener las mismas divisiones de datos en diferentes computadoras y ejecuciiones, fijamos este parámetro en un mismo valor, conocido como valor semilla. En nuestros ejercicios utilizaremos la semilla 0.

````python
XTrain, XTest, yTrain, yTest = train_test_split(XScaled, y, test_size = 0.3, random_state=0)

print(XTrain.shape , yTrain.shape)
print(XTest.shape , yTest.shape)
````

### 5.2) Inicialización y Entrenamiento del Modelo:

````python
logReg1 = LogisticRegression(multi_class='ovr',solver='lbfgs', random_state=0) 
logReg1.fit(XTrain, yTrain)

print('Coeficientes: ', logReg1.coef_)
print('Intercepto: ', logReg1.intercept_)
print('Clases: ', logReg1.classes_)
````

También podemos crear el modelo desdse **statsmodels** para tener la posibilidad de acceder al analisis estadístico de la prueba de hipótesis del modelo y la significancia de sus coeficientes.

````python
logReg2=sm.Logit(yTrain,XTrain,random_state=0)
logModel2=logReg2.fit()

print('\n** Parametros **')
print(logModel2.params)
print('\n** p-values **')
print(logModel2.pvalues)
print('\n** Resumen General **')
print(logModel2.summary2())
````

De acuerdo a los p-values obtenidos en la columna ```P>|z|``` para los predictores ```x1('student') y x3('income')```, podemos concluir que **es muy poco probable observar una asociación substancial entre los predictores 'student' e 'income' y la variable de respuesta a predecir 'default'** en el modelo de regresión logística en análisis.

### 5.3) Predicción

````python
yHat = logReg1.predict(XTest)

print('yTest: Default YES = %d , NO = %d' % (sum(yTest==1), sum(yTest==0)))
print('yHat:  Default YES = %d , NO = %d' % (sum(yHat==1), sum(yHat == 0)))
````

## 6) Evaluación del Modelo
### 6.1) Métricas

En tareas o algoritmos de clasificación utilizamos los siguientes conceptos para evaluar los resultados obtenidos: <br>

- (TP): Numero de Verdaderos Positivos obtenidos
- (TN): Numero de Verdaderos Negativos obtenidos
- (FP): Numero de Falsos Positivos obtenidos
- (FN): Numero de Falsos Negativos obtenidos

Con estos términos, recordemos las definiciones de las métricas de evaluación de un algoritmo de ML:

- Accuracy ***(Exactitud)*** = TP + TN / (TP + TN + FP + FN)
- Precision ***(Precision)*** = TP / (TP + FP)
- Recall ***(Sensitividad / Verdaderos Positivos)*** = TP / (TP + FN) <br>

A continuación calculamos estas métricas para nuestro modelo:

````python
#Exactitud (Accuracy) del Modelo
print('Exactitud (Accuracy): {:.2f}'.format(metrics.accuracy_score(yTest, yHat)))
print('Precisión: {:.2f}'.format(metrics.precision_score(yTest, yHat)))
print('Recall: {:.2f}'.format(metrics.recall_score(yTest, yHat)))
````

### 6.2) Matriz de Confusión
Es un resumen de conteos de los resultados de la clasificación. En nuestro caso este resultado se muestra en una matriz de 2 filas y 2 columnas, donde cada (fila,columna) representan:

- (1,1) indica cuantos ```Defaults Falsos (0) ``` fueron clasificados correctamente como ```Falsos(0)```
- (1,2) indica cuantos ```Defaults Falsos (0) ``` fueron clasificados erroneamente como ```Verdaderos(1)```
- (2,1) indica cuantos ```Defaults Verdaderos (1) ``` fueron clasificados erroneamente como ```Falsos(0)```
- (2,2) indica cuantos ```Defaults Verdaderos (1) ``` fueron clasificados correctamente como ```Verdaderos(1)```

````python
#Matriz de Confusion
MC = metrics.confusion_matrix(yTest, yHat)
print("\n** Matriz de Confusion **")
print(MC)
````

````python
plt.figure()
boxplot1 = sb.boxplot(data=defaultData,x="default", y="balance")
display(boxplot1.figure)
````

````python
plt.figure()
p1 = sb.heatmap(MC, annot=True, fmt=".2f", linewidths=.5,cmap = 'Greens_r');
plt.ylabel('yTest');
plt.xlabel('yHat');
titulo = 'Accuracy: {0:.2f}'.format(metrics.accuracy_score(yTest, yHat))
plt.title(titulo, size = 14);
display(p1.figure)
````

### 6.3) Curva ROC-AUC  

Representa el desempeño del clasificador en términos de la relación entre el número de Falsos Positivos vs. el número de Verdaderos Positivos. **El área que se forma bajo esta curva es un indicador de la exactitud del clasificador: debe ser cercana a 1 para indicar un buen desempeño.** <br>

- Mientras más rápido converge esta curva hacia 1, la clasificación es mejor. <br>
- Si la curva se encuentra por debajo de la recta ```y=x``` (AUC <= 0.5), el clasificador no es mejor que el azar (es trivial).

````python
#SP: Grafico con plt.plot

#Calculo de FPR (falsos positivos) y TPR (verdaderos positivos):
scores = logReg1.predict_proba(XTest)
fpr, tpr, threshold = metrics.roc_curve(yTest, scores[:,1])

#Calculo de ROC-AUC
aucROC = metrics.auc(fpr, tpr)

fig1 = plt.figure()
plt.plot(fpr,tpr, 'b', label = 'AUC = {:0.2f}'.format(aucROC))
plt.plot([0, 1], [0, 1],'r--', lw=3, linestyle='--')
plt.legend(loc = 'lower right')
plt.xlim([-0.01, 1.00])
plt.ylim([-0.01, 1.01])
plt.ylabel('Ratio Verdaderos Positivos', fontsize=14)
plt.xlabel('Ratio Falsos Positivos', fontsize=14)
plt.title('Curva ROC: Receiver Operating Characteristic', fontsize=16)
display(fig1)
````
